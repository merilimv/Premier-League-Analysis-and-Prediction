```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
results = pd.read_csv("results.csv", delimiter=",")

results.head()
```

```python
mult_train = results.loc[(results["Type"] == "Mult") & (results["Set"] == "train")].drop("Type", axis = 1)
mult_test = results.loc[(results["Type"] == "Mult") & (results["Set"] == "test")].drop("Type", axis = 1)

mult_train
```

```python
bin_train = results.loc[(results["Type"] == "Bin") & (results["Set"] == "train")].drop("Type", axis = 1)
bin_test = results.loc[(results["Type"] == "Bin") & (results["Set"] == "test")].drop("Type", axis = 1)

bin_train
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 10))

# Plotagem do gráfico de barras com Seaborn
cores = sns.color_palette("mako", 3)  # Lista de cores
sns.barplot(data=bin_train, x="Metric", y="Value", hue="Model", errorbar=None, ax=ax[0], linewidth=2, palette=cores)

sns.set(style="darkgrid")
sns.despine()
# Configurações do gráfico
ax[0].set_xlabel('Métrica', fontsize=16)
ax[0].set_ylabel('Valor da Métrica', fontsize=16)
ax[0].set_title('Comparação dos Modelos dos Dados de Treino', fontsize = 23)
ax[0].legend(title='Modelo', fontsize=14)
ax[0].tick_params(axis='x', rotation=20, labelsize=14)

sns.barplot(data=bin_test, x="Metric", y="Value", hue="Model", errorbar=None, ax=ax[1], linewidth=2, palette=cores)

sns.set(style="darkgrid")
sns.despine()
ax[1].set_xlabel('Métrica', fontsize=16)
ax[1].set_ylabel('Valor da Métrica', fontsize=16)
ax[1].set_title('Comparação dos Modelos dos Dados de Treino', fontsize = 23)
ax[1].legend(title='Modelo', fontsize=14)
ax[1].tick_params(axis='x', rotation=20, labelsize=14)

# Modificar o ax[1]
ax[1].set_ylim(0, 1)  # Define o valor máximo de y para o segundo ax
ax[0].set_ylim(0, 1)  # Define o valor máximo de y para o segundo ax

plt.tight_layout()

# Exibe o gráfico
plt.show()

fig.savefig("./gráficos/Results/binaryclass.png")

```

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 10))

# Plotagem do gráfico de barras com Seaborn
cores = sns.color_palette("mako", 3)  # Lista de cores
sns.barplot(data=mult_train, x="Metric", y="Value", hue="Model", errorbar=None, ax=ax[0], linewidth=2, palette=cores)
sns.set(style="darkgrid")
sns.despine()
# Configurações do gráfico
ax[0].set_xlabel('Métrica', fontsize=16)
ax[0].set_ylabel('Valor da Métrica', fontsize=16)
ax[0].set_title('Comparação dos Modelos dos Dados de Treino', fontsize = 23)
ax[0].legend(title='Modelo', fontsize=14)
ax[0].tick_params(axis='x', rotation=20, labelsize=14)

sns.barplot(data=mult_test, x="Metric", y="Value", hue="Model", errorbar=None, ax=ax[1], linewidth=2, palette=cores)
sns.set(style="darkgrid")
sns.despine()
ax[1].set_xlabel('Métrica', fontsize=16)
ax[1].set_ylabel('Valor da Métrica', fontsize=16)
ax[1].set_title('Comparação dos Modelos dos Dados de Treino', fontsize = 23)
ax[1].legend(title='Modelo', fontsize=14)
ax[1].tick_params(axis='x', rotation=20, labelsize=18)

# Modificar o ax[1]
ax[1].set_ylim(0, 0.7)  # Define o valor máximo de y para o segundo ax
ax[0].set_ylim(0, 0.7)  # Define o valor máximo de y para o segundo ax

# Exibe o gráfico
plt.tight_layout()
plt.show()

fig.savefig("./gráficos/Results/multiclass.png")


```
